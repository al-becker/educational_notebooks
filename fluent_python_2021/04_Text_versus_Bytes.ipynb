### encoding and decoding

```python
s = 'café'
len(s)
```

```python
b = s.encode('utf8')
b
```

```python
len(b)
```

```python
b.decode('utf8')
```

```python
b[0]
```

```python
b[:1]
```

```python
cafe_arr = bytearray(b)
cafe_arr
```

```python
cafe_arr[-1:]
```

```python
for codec in ['latin_1', 'utf_8', 'utf_16']:
    print(codec, 'El Niño'.encode(codec), sep='\t')
```

```python
city.encode('utf_16')
```
